# 범죄자_범행시_연령_4개년

```python
age = pd.read_csv("범죄자_범행시_연령_4개년.csv",encoding = 'euc-kr')
age.replace('-','0',inplace=True) # 결측치를 0으로 변경
```

#### 전처리1

```python
# 죄종별(2)의 소계에 해당하는 row 제거 
# 조건1: 죄종별(1)에 해당하는 값이 죄종별(2) 하나 뿐이라면 소계를 죄종별(1)값으로 변경
# 조건2: 죄종별(1)에 해당하는 값이 2개 이상이라면 소계는 제거

cond = (age['죄종별(2)']=='소계')
age.loc[cond,'죄종별(2)']=age.loc[cond,'죄종별(1)'] # 죄종별(2)가 소계인 경우 죄종별(1)값으로 대체

cond1 = age[age['죄종별(1)'] != age['죄종별(2)']].index # 죄종별(1)과 (2)가 같지 않다면 선택
cond2 = age[(~(age['죄종별(1)'].duplicated(keep=False)) & (age['죄종별(1)'] == age['죄종별(2)']))].index # 죄종별(1)과 (2)가 같지만 하나뿐이라면 선택
final_index = cond1.to_list() + cond2.to_list()
age = age.loc[final_index].sort_index().reset_index(drop=True)
#age
```

```python
age19=['죄종별(1)','죄종별(2)']
age20=['죄종별(1)','죄종별(2)']
age21=['죄종별(1)','죄종별(2)']
age22=['죄종별(1)','죄종별(2)']
for i in age.columns:
    if "2019." in i:
        age19.append(i)
    elif "2020." in i:
        age20.append(i)
    elif "2021." in i:
        age21.append(i)
    elif "2022." in i:
        age22.append(i)

age19 = age19[0:2]+age19[3:8]+age19[9:]
age19 = age[age19]
age19.columns = age19.iloc[1]
age19 = age19.drop(labels=[0,1,2],axis=0)
age19= age19.reset_index(drop=True)
age19.rename(columns={'소계':'미상'},inplace=True)

age20 = age20[0:2]+age20[3:8]+age20[9:]
age20 = age[age20]
age20.columns = age20.iloc[1]
age20 = age20.drop(labels=[0,1,2],axis=0)
age20= age20.reset_index(drop=True)
age20.rename(columns={'소계':'미상'},inplace=True)

age21 = age21[0:2]+age21[3:8]+age21[9:]
age21 = age[age21]
age21.columns = age21.iloc[1]
age21 = age21.drop(labels=[0,1,2],axis=0)
age21= age21.reset_index(drop=True)
age21.rename(columns={'소계':'미상'},inplace=True)

age22 = age22[0:2]+age22[3:8]+age22[9:]
age22 = age[age22]
age22.columns = age22.iloc[1]
age22 = age22.drop(labels=[0,1,2],axis=0)
age22= age22.reset_index(drop=True)
age22.rename(columns={'소계':'미상'},inplace=True)
```

# 소년_범죄자_교육정도_4개년

```python
edu = pd.read_csv("소년범죄자_교육정도_4개년.csv",encoding = 'euc-kr')
edu.replace('-','0',inplace=True) # 결측치를 0으로 변경
edu.iloc[0:20,0:3]
```

#### 전처리2

```python
cond = (edu['범죄별(3)']=='소계')
edu.loc[cond,'범죄별(3)']=edu.loc[cond,'범죄별(2)'] # 범죄별(3)가 소계인 경우 범죄별(2)값으로 대체

cond1 = edu[edu['범죄별(2)'] != edu['범죄별(3)']].index # 범죄별(2)와 (3)가 같지 않다면 선택
cond2 = edu[(~(edu['범죄별(2)'].duplicated(keep=False)) & (edu['범죄별(2)'] == edu['범죄별(3)']))].index 
final_index = cond1.to_list() + cond2.to_list()
edu = edu.loc[final_index].sort_index().reset_index(drop=True)

edu = edu.drop('범죄별(1)',axis=1) # 범죄별(1) 의미 없음.
edu
```

```python
edu19=['범죄별(2)','범죄별(3)']
edu20=['범죄별(2)','범죄별(3)']
edu21=['범죄별(2)','범죄별(3)']
edu22=['범죄별(2)','범죄별(3)']
for i in edu.columns:
    if "2019." in i:
        edu19.append(i)
    elif "2020." in i:
        edu20.append(i)
    elif "2021." in i:
        edu21.append(i)
    elif "2022." in i:
        edu22.append(i)

edu19 = edu[edu19]
edu19.columns = edu19.iloc[0]+''+edu19.iloc[1]

edu19 = edu19.drop(labels=[0,1],axis=0)
edu19= edu19.reset_index(drop=True)
edu19.rename(columns={'범죄별(2)범죄별(2)':'범죄별(2)','범죄별(3)범죄별(3)':'범죄별(3)'},inplace=True)

edu20 = edu[edu20]
edu20.columns = edu20.iloc[0]+''+edu20.iloc[1]
edu20 = edu20.drop(labels=[0,1],axis=0)
edu20= edu20.reset_index(drop=True)
edu20.rename(columns={'범죄별(2)범죄별(2)':'범죄별(2)','범죄별(3)범죄별(3)':'범죄별(3)'},inplace=True)

edu21 = edu[edu21]
edu21.columns = edu21.iloc[0]+''+edu21.iloc[1]
edu21 = edu21.drop(labels=[0,1],axis=0)
edu21= edu21.reset_index(drop=True)
edu21.rename(columns={'범죄별(2)범죄별(2)':'범죄별(2)','범죄별(3)범죄별(3)':'범죄별(3)'},inplace=True)

edu22 = edu[edu22]
edu22.columns = edu22.iloc[0]+''+edu22.iloc[1]
edu22 = edu22.drop(labels=[0,1],axis=0)
edu22= edu22.reset_index(drop=True)
edu22.rename(columns={'범죄별(2)범죄별(2)':'범죄별(2)','범죄별(3)범죄별(3)':'범죄별(3)'},inplace=True)
```

#### 전처리3

```python
cond = (sex_mental['범죄별(3)']=='소계')
sex_mental.loc[cond,'범죄별(3)']=sex_mental.loc[cond,'범죄별(2)'] # 범죄별(3)가 소계인 경우 범죄별(2)값으로 대체

cond1 = sex_mental[sex_mental['범죄별(2)'] != sex_mental['범죄별(3)']].index # 범죄별(2)와 (3)가 같지 않다면 선택
cond2 = sex_mental[(~(sex_mental['범죄별(2)'].duplicated(keep=False)) & (sex_mental['범죄별(2)'] == sex_mental['범죄별(3)']))].index 
final_index = cond1.to_list() + cond2.to_list()
sex_mental = sex_mental.loc[final_index].sort_index().reset_index(drop=True)

sex_mental = sex_mental.drop('범죄별(1)',axis=1) # 범죄별(1) 의미 없음.

for i,data in enumerate(sex_mental.loc[0]): # 정신장애 분류
    if data == '정신장애':
        sex_mental.iloc[0,i]=sex_mental.iloc[1,i]
sex_mental=sex_mental.drop(labels=[1],axis=0).reset_index(drop=True)

sex_mental
```
